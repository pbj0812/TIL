```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# 위에 있는 것을 써야 그림이 바로 나옴
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
# Functional Method
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Titile')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(y, x)
```

```python
# object-oriented method
fig = plt.figure()
```

```python
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
ax1.plot(x, y)
ax2.plot(y, x)

ax1.set_title('Larger Plot')
ax2.set_title('smaqller Plot')

fig
```

```python
fig = plt.figure(figsize = (8, 1), dpi = 60)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
fig
```

```python
fig.savefig('my_pic.png', dpi = 200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label = 'X squared')
ax.plot(x, x ** 3, label = 'X cubed')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )

ax.plot( x, y, color = 'purple', lw = 2, ls = '--' )

ax.set_xlim( [0, 1] )
ax.set_ylim( [0, 2] )
print('x : ', x)
print('y : ', y)
```
